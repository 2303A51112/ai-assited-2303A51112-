Task 1: Merge Sort (Sorting)

```python
def merge_sort(arr):
    """
    Merge Sort Algorithm

    Time Complexity:
    Best Case: O(n log n)
    Average Case: O(n log n)
    Worst Case: O(n log n)

    Space Complexity: O(n)
    """

    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

    return arr


# Test
data = [38, 27, 43, 3, 9, 82, 10]
print("Sorted array:", merge_sort(data))
```

Task 2: Binary Search

```python
def binary_search(arr, target):
    """
    Binary Search Algorithm

    Best Case: O(1)
    Average Case: O(log n)
    Worst Case: O(log n)

    Works only on sorted arrays.
    """

    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1


# Test
arr = [2, 5, 8, 12, 16, 23, 38]
print("Element found at index:", binary_search(arr, 16))
```

Task 3: Smart Healthcare Appointment System

```python
appointments = [
    (101, "Ravi", "Dr. Kumar", "10:30", 500),
    (102, "Anita", "Dr. Rao", "09:45", 300),
    (103, "Vikram", "Dr. Sharma", "11:15", 700)
]

def search_appointment(appointments, target_id):
    for appt in appointments:
        if appt[0] == target_id:
            return appt
    return "Not Found"

def sort_by_fee(appointments):
    return sorted(appointments, key=lambda x: x[4])

print(search_appointment(appointments, 102))
print(sort_by_fee(appointments))
```

Task 4: Railway Ticket Reservation System

```python
tickets = [
    (201, "Rahul", "12711", "S5-23", "2026-03-10"),
    (202, "Priya", "12645", "S2-11", "2026-02-28"),
    (203, "Arjun", "12833", "S1-05", "2026-03-05")
]

def search_ticket(tickets, ticket_id):
    for t in tickets:
        if t[0] == ticket_id:
            return t
    return "Ticket Not Found"

def sort_by_date(tickets):
    return sorted(tickets, key=lambda x: x[4])

print(search_ticket(tickets, 202))
print(sort_by_date(tickets))
```

Task 6: Online Movie Streaming Platform

```python
movies = [
    (1, "Inception", "Sci-Fi", 8.8, 2010),
    (2, "Avatar", "Action", 7.9, 2009),
    (3, "Interstellar", "Sci-Fi", 8.6, 2014)
]

def search_movie(movies, movie_id):
    for m in movies:
        if m[0] == movie_id:
            return m
    return "Movie Not Found"

def sort_by_rating(movies):
    return sorted(movies, key=lambda x: x[3], reverse=True)

print(search_movie(movies, 1))
print(sort_by_rating(movies))
```

Task 7: Smart Agriculture Crop Monitoring

```python
crops = [
    (1, "Rice", 45, 30, 200),
    (2, "Wheat", 35, 28, 180),
    (3, "Corn", 50, 32, 220)
]

def search_crop(crops, crop_id):
    for c in crops:
        if c[0] == crop_id:
            return c
    return "Crop Not Found"

def sort_by_moisture(crops):
    return sorted(crops, key=lambda x: x[2])

print(search_crop(crops, 2))
print(sort_by_moisture(crops))
```

Task 8: Airport Flight Management System

```python
flights = [
    (501, "IndiGo", "10:00", "12:00", "On Time"),
    (502, "Air India", "09:30", "11:45", "Delayed"),
    (503, "SpiceJet", "11:15", "13:20", "On Time")
]

def search_flight(flights, flight_id):
    for f in flights:
        if f[0] == flight_id:
            return f
    return "Flight Not Found"

def sort_by_departure(flights):
    return sorted(flights, key=lambda x: x[2])

print(search_flight(flights, 502))
print(sort_by_departure(flights))
```
