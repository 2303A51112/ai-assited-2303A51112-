```python
# Task 1: Mutable Default Argument
def add_item(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items

print("Task 1 Output:")
print(add_item(1))
print(add_item(2))
print("-" * 30)


# Task 2: Floating-Point Precision Error
import math

def check_sum():
    return math.isclose(0.1 + 0.2, 0.3)

print("Task 2 Output:")
print(check_sum())
print("-" * 30)


# Task 3: Recursion Error (Missing Base Case)

def countdown(n):
    if n < 0:   # base case
        return
    print(n)
    countdown(n - 1)

print("Task 3 Output:")
countdown(5)
print("-" * 30)


# Task 4: Dictionary Key Error
def get_value():
    data = {"a": 1, "b": 2}
    return data.get("c", "Key not found")

print("Task 4 Output:")
print(get_value())
print("-" * 30)


# Task 5: Infinite Loop (Wrong Condition)
def loop_example():
    i = 0
    while i < 5:
        print(i)
        i += 1

print("Task 5 Output:")
loop_example()
print("-" * 30)


# Task 6: Unpacking Error
a, b, _ = (1, 2, 3)

print("Task 6 Output:")
print(a, b)
print("-" * 30)


# Task 7: Mixed Indentation
def func():
    x = 5
    y = 10
    return x + y

print("Task 7 Output:")
print(func())
print("-" * 30)


# Task 8: Import Error
import math

print("Task 8 Output:")
print(math.sqrt(16))
print("-" * 30)
```
