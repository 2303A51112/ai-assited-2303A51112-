Task 1: Transparency in Algorithm Optimization

```python
import math

def is_prime_optimized(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

Task 2: Transparency in Recursive Algorithms (Fibonacci)

```python
def fibonacci(n):
    # Base case 1: Fibonacci of 0 is 0
    if n == 0:
        return 0

    # Base case 2: Fibonacci of 1 is 1
    if n == 1:
        return 1

    # Recursive call: sum of previous two Fibonacci numbers
    return fibonacci(n - 1) + fibonacci(n - 2)
```

Task 3: Transparency in Error Handling

```python
def read_file(filename):
    try:
        file = open(filename, 'r')
        data = file.read()
        file.close()
        return data

    except FileNotFoundError:
        print("Error: File not found.")

    except PermissionError:
        print("Error: Permission denied.")

    except Exception as e:
        print("Unexpected error:", e)
```
