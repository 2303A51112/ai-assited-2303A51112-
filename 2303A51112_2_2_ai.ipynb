```python
Task 1: Cleaning Sensor Data
```

```python
def remove_negative_values(data):
    return [value for value in data if value >= 0]

sensor_data = [12, -3, 7, -1, 0, 25]
cleaned_data = remove_negative_values(sensor_data)

print("Before cleaning:", sensor_data)
print("After cleaning:", cleaned_data)
```

Task 2: String Character Analysis

```python
def analyze_string(text):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0

    for char in text:
        if char.isdigit():
            digit_count += 1
        elif char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count, digit_count


text = "Hello123"
v, c, d = analyze_string(text)

print("Vowels:", v)
print("Consonants:", c)
print("Digits:", d)
```

Task 3: Palindrome Check

```python
def is_palindrome(s):
    s = s.lower()
    return s == "".join(reversed(s))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```
